```python
# Dependencies
import os 
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress 

#Reading in Files
cdc_df = pd.read_csv("Resource/VSRR_Provisional_Drug_Overdose_Death_Counts_2015_2021.csv", thousands=',')
care_df = pd.read_csv("Resource/Medicare_Opioid_Prescribing_Data_2013_2019.csv", low_memory = False)
caid_df = pd.read_csv("Resource/Medicaid_Opioid_Prescribing_data_2013_2019.csv", low_memory = False)
```

```python
# Filtering Medcaid dropping year 2013 & 2014 and filtering for all plan types
caid_year = caid_df.loc[(caid_df["Year"] != 2013) & (caid_df["Year"] != 2014)]
caid_clean = caid_year.loc[caid_year["Plan_Type"]=="All"]
# Reset the index values for the dataframe
caid_clean = caid_clean.reset_index(drop=True)
caid_clean
```

```python
# Changing column name to state for merging later
caid_clean = caid_clean.rename(columns={"Geo_Desc":"State"})

caid_clean
```

```python
# Filtering Medicaid Data by Breakout Type and Perscription Geographic Levels for National or State levels
care_year = care_df.loc[(care_df["Year"] != 2013) & (care_df["Year"] != 2014)]
care_clean = care_year.loc[(care_year["Breakout_Type"]=="Totals") &  (care_year["Prscrbr_Geo_Lvl"]=="National") | 
                           (care_year["Prscrbr_Geo_Lvl"]=="State")  &  (care_year["Breakout"]=="Overall") & 
                           (care_df["Prscrbr_Geo_Desc"] != 'American Samoa')& (care_df["Prscrbr_Geo_Desc"] != 'Guam') & 
                           (care_df["Prscrbr_Geo_Desc"] != 'American Samoa') & (care_df["Prscrbr_Geo_Desc"] != 'Northern Mariana Islands') & 
                           (care_df["Prscrbr_Geo_Desc"] != 'Puerto Rico') & (care_df["Prscrbr_Geo_Desc"] != 'Virgin Islands')]

# Reset the index values for the dataframe
care_clean = care_clean.reset_index(drop=True)
care_clean
```

```python
#print(care_clean['RUCA_Cd'].unique())
```

```python
# drop column RUCA_CD with nan values
del care_clean["RUCA_Cd"]
```

```python
# Changing column name to state for merging late
care_clean = care_clean.rename(columns={"Prscrbr_Geo_Desc":"State"})
care_clean
```

```python
# Dropping years 2020 & 2021 and filtering indicator column 
cdc_year = cdc_df.loc[(cdc_df["Year"] != 2020) & (cdc_df["Year"] != 2021) & (cdc_df["State"] != 'YC')]
cdc_numofdrugdeathsindicator = cdc_year.loc[(cdc_year["Indicator"]=="Number of Drug Overdose Deaths")]

# Reset the index values for the dataframes
cdc_numofdrugdeathsindicator = cdc_numofdrugdeathsindicator.reset_index(drop=True)
cdc_numofdrugdeathsindicator

# Changing column names for merging later
cdc_numofdrugdeathsindicator = cdc_numofdrugdeathsindicator.rename(columns={"State" : "State ID", "State Name" : "State"})
cdc_numofdrugdeathsindicator['Data Value'] = pd.to_numeric(cdc_numofdrugdeathsindicator['Data Value'], errors = 'coerce')

# Dropping years 2020 & 2021 and filtering indicator column 
cdc_year = cdc_df.loc[(cdc_df["Year"] != 2020) & (cdc_df["Year"] != 2021) & (cdc_df["State"] != 'YC')]
cdc_numofdrugdeathsindicator = cdc_year.loc[(cdc_year["Indicator"]=="Number of Drug Overdose Deaths")]

# Changing column names for merging later
cdc_numofdrugdeathsindicator = cdc_numofdrugdeathsindicator.rename(columns={"State" : "State ID", "State Name" : "State"})
cdc_numofdrugdeathsindicator['Data Value'] = pd.to_numeric(cdc_numofdrugdeathsindicator['Data Value'], errors = 'coerce')

# Group by State ID, Year and sum to find annual total death
cdc_numofdrugdeathsaggregated=cdc_numofdrugdeathsindicator.groupby(['State ID', 'Year'])["Data Value"].sum().reset_index(name ='Total')


cdc_numofdrugdeathsaggregated
```

```python
# exclude State ID = 'US'
cdc_numofdrugdeathsaggregated = cdc_numofdrugdeathsaggregated.loc[(cdc_numofdrugdeathsaggregated["State ID"] != 'US')] 
cdc_numofdrugdeathsaggregated
```

```python
# cdc state with max deaths 

cdc_max = cdc_numofdrugdeathsaggregated.loc[cdc_numofdrugdeathsaggregated.groupby('Year')['Total'].idxmax()]

cdc_max.head()
```

```python
# create series for total death, x_values
Total_Series = cdc_max.iloc[:,2]
x_values = cdc_max.iloc[:,0]

# Plot the figure.
x_label2 =['2015--', '2016--', '2017--', '2018--', '2019--']
plt.figure(figsize=(12, 8))
ax = Total_Series.plot(kind="bar")
ax.set_title("State with Max Opioid Death")
ax.set_xlabel("State")
ax.set_ylabel("Total Death")
ax.set_xticklabels(x_label2 + x_values, rotation = 45)

# Place label on top on the chart
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
    
# save figure
plt.savefig("Image/StateWithMaxOpioidDeath.png")
```

```python
# cdc state with min deaths 
cdc_min = cdc_numofdrugdeathsaggregated.loc[cdc_numofdrugdeathsaggregated.groupby('Year')['Total'].idxmin()]

cdc_min.head()
```

```python
# create series for total death, x_values
Total_Series_min = cdc_min.iloc[:,2]
x_values_min = cdc_min.iloc[:,0]

# Plot the figure.
x_label3 =['2015--', '2016--', '2017--', '2018--', '2019--']
plt.figure(figsize=(12, 8))
ax= Total_Series_min.plot(kind="bar")
ax.set_title("State with Min Opioid Death")
ax.set_xlabel("State")
ax.set_ylabel("Total Death")
ax.set_xticklabels(x_label3 + x_values_min, rotation = 45)

# Place label on top on the chart
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
    
# save figure
plt.savefig("Image/StateWithMinOpioidDeath.png")
```
