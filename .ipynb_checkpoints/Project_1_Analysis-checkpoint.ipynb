```python
# Dependencies
import os 
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import gmaps
from scipy.stats import linregress 

# Import API key
from api_keys import g_key

# Import Dependencies for Heatmap of Overdose Deaths
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

# Import Dependencies for Heatmap of Overdose Deaths
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets

#Reading in Files
cdc_df = pd.read_csv("Resource/VSRR_Provisional_Drug_Overdose_Death_Counts_2015_2021.csv", thousands=',')
care_df = pd.read_csv("Resource/Medicare_Opioid_Prescribing_Data_2013_2019.csv", low_memory = False)
caid_df = pd.read_csv("Resource/Medicaid_Opioid_Prescribing_data_2013_2019.csv", low_memory = False)

#preview cdc file
cdc_df.head()
```

# Data Cleaning

```python
# Medicaid data cleanup
# Filtering Medcaid dropping year 2013 & 2014 and filtering for all plan types
caid_clean = caid_df.loc[(caid_df["Year"] != 2013) & (caid_df["Year"] != 2014)]
caid_clean = caid_clean.loc[caid_clean["Plan_Type"]=="All"]

# Reset the index values for the dataframe
caid_clean = caid_clean.reset_index(drop=True)

# Changing column name to state for merging late
caid_clean = caid_clean.rename(columns={"Geo_Desc":"State"})
caid_clean
```

```python
# Medicare data 
# Filtering Medicare data by dropping 2013 and 2014 data, Breakout Type, and Perscription Geographic Levels for National or State levels
care_year = care_df.loc[(care_df["Year"] != 2013) & (care_df["Year"] != 2014)]
care_clean = care_year.loc[(care_year["Breakout_Type"]=="Totals") &  (care_year["Prscrbr_Geo_Lvl"]=="National") | (care_year["Prscrbr_Geo_Lvl"]=="State")  &  (care_year["Breakout"]=="Overall") & (care_df["Prscrbr_Geo_Desc"] != 'American Samoa')& (care_df["Prscrbr_Geo_Desc"] != 'Guam') & (care_df["Prscrbr_Geo_Desc"] != 'American Samoa') & (care_df["Prscrbr_Geo_Desc"] != 'Northern Mariana Islands') & (care_df["Prscrbr_Geo_Desc"] != 'Puerto Rico') & (care_df["Prscrbr_Geo_Desc"] != 'Virgin Islands')]

# Reset the index values for the dataframe
care_clean = care_clean.reset_index(drop=True)

# Changing column name to state for merging late
care_clean = care_clean.rename(columns={"Prscrbr_Geo_Desc":"State"})
care_clean
```

```python
# CDC data
# Dropping years 2020 & 2021 and filtering indicator column 
cdc_clean = cdc_df.loc[(cdc_df["Year"] != 2020) & (cdc_df["Year"] != 2021) & (cdc_df["State"] != 'YC')]
cdc_clean
```

```python
# Create dataframe without these state values:
drop = ["National", "District of Columbia", "United States"]

cdc_clean2 = cdc_clean.loc[cdc_clean["State"].isin(drop)==False]
cdc_clean2
```

# Aggregating Data

## Number of total deaths 

```python
# Finding number of deaths and reset the index values
cdc_deaths = cdc_clean2.loc[(cdc_clean2["Indicator"]=="Number of Deaths")].reset_index(drop=True)

cdc_deaths
```

```python
# Changing column names for merging later
cdc_deaths = cdc_deaths.rename(columns={"State" : "State ID", "State Name" : "State"})
cdc_deaths['Data Value'] = pd.to_numeric(cdc_deaths['Data Value'], errors = 'coerce')
cdc_deathsagg = cdc_deaths.groupby(['State ID', 'Year']).agg({'Data Value' : ['sum']})

cdc_deathsagg
```

## Number of drug overdose deaths 

```python
# Dropping years 2020 & 2021, filtering indicator column, resetting index
cdc_drugdeaths = cdc_clean2.loc[(cdc_clean2["Indicator"]=="Number of Drug Overdose Deaths")].reset_index(drop=True)
cdc_drugdeaths
```

```python
# Changing column names for merging later
cdc_drugdeaths = cdc_drugdeaths.rename(columns={"State" : "State ID", "State Name" : "State"})
cdc_drugdeaths['Data Value'] = pd.to_numeric(cdc_drugdeaths['Data Value'], errors = 'coerce')

# Creating aggregated dataframe, grouped by State ID and Year
cdc_aggdrugdeaths = cdc_drugdeaths.groupby(['State ID', 'Year']).agg({'Data Value' : ['sum']})
cdc_aggdrugdeaths
```

## Number of opioid deaths 

```python
# Create dataframe of opioid-specific deaths & reset index
cdc_opioiddeaths = cdc_clean2.loc[(cdc_clean["Indicator"]=="Opioids (T40.0-T40.4,T40.6)")].reset_index(drop=True)
cdc_opioiddeaths                                             
```

```python
# Changing column names for merging later
cdc_opioiddeaths = cdc_opioiddeaths.rename(columns={"State" : "State ID", "State Name" : "State"})
cdc_opioiddeaths['Data Value'] = pd.to_numeric(cdc_opioiddeaths['Data Value'], errors = 'coerce')
cdc_aggopioiddeaths = cdc_opioiddeaths.groupby(['State ID', 'Year']).agg({'Data Value' : ['sum']})

cdc_aggopioiddeaths
                                
```

```python
# Merging Medicaid & Medicare data with an Outer Join
carecaid = pd.merge(care_clean, caid_clean, how = "outer", on = ["Year", "State"])
carecaid
```

# Google Heatmap

```python
# Creating Google Heatmaps of Overdose Deaths by State

# Import Google API key for configuration
gmaps.configure(api_key = g_key)

# Importing mean population center coordinates for the 50 states so that data points coincide with centers
state_lat_long_df = pd.read_csv("Resource/CenPop2020_Mean_ST.csv")
state_lat_long_df = state_lat_long_df.loc[state_lat_long_df['STNAME'] != 'Puerto Rico']
state_coords = state_lat_long_df[['LATITUDE', 'LONGITUDE']]

# Filtering out cdc_numofdrugdeathsaggregated to exclude the total US deaths
# The year had to be manually changed to get heatmaps of overdose deaths 
# for the analysed years (2015, 2016, 2017, 2018, & 2019)
num_deaths_df = cdc_deathsagg.reset_index().droplevel(level = 1, axis = 1).rename(columns = {'Data Value':'Sum'})
num_deaths_year_df = num_deaths_df.loc[(num_deaths_df['Year'] == 2019) & (num_deaths_df['State ID'] != 'US')]
num_deaths_year = num_deaths_year_df['Sum']

# Layout of the map
layout = {'width': '1000px', 'height': '500px', 'border': '1px solid black', 'paddling': '1px'}
# Centered the figure on the mean population center of the United States including Alaska and Hawaii
# https://www.census.gov/geographies/reference-files/time-series/geo/centers-population.html
center = [44.97, -103.77]

# Generating the heatmap
fig = gmaps.figure(map_type = 'HYBRID', layout = layout, center = center, zoom_level = 3)
heat = gmaps.heatmap_layer(state_coords, weights = num_deaths_year, dissipating = False, \
                           max_intensity = 5e4, point_radius = 3)

fig.add_layer(heat)
fig
```

```python
#Build out Table for Count of Drug Deaths by State and Year
cdc_drugdeaths['Year_State'] = cdc_drugdeaths['Year'].map(str) +'_'+ cdc_drugdeaths['State'].map(str)
cdc_drugdeaths = cdc_drugdeaths.rename(columns={'Data Value':'CountofDrugDeaths'})

#Group By Year/State, Sum all integer columnsm, then drop the unnecessary columns
DrugDeathsByStateAndYear = cdc_drugdeaths.groupby(['Year_State']).sum()
DrugDeathsByStateAndYear = DrugDeathsByStateAndYear.drop(columns=['Year','Percent Pending Investigation','Predicted Value'])

#Build out Table for Count of Drug Deaths by State and Year
cdc_deaths['Year_State'] = cdc_deaths['Year'].map(str)+'_'+ cdc_deaths['State'].map(str)
cdc_deaths = cdc_deaths.rename(columns={'Data Value':'CountofDeaths'})

#Group By Year/State, Sum all integer columnsm, then drop the unnecessary columns
DeathsByStateAndYear = cdc_deaths.groupby(['Year_State']).sum()
DeathsByStateAndYear = DeathsByStateAndYear.drop(columns=['Year','Percent Pending Investigation','Predicted Value'])

#Merge Drug Deaths and Overall Deaths into single DataFrame, Reset Inxes, Filter Columns, Calculate Ratio
DrugDeathRatio = pd.merge(DrugDeathsByStateAndYear, DeathsByStateAndYear, how = "outer", on = ['Year_State'])
DrugDeathRatio = DrugDeathRatio.reset_index()
DrugDeathRatioFinal = DrugDeathRatio.filter(['Year_State','CountofDrugDeaths','CountofDeaths'])
DrugDeathRatioFinal['DrugDeathRatio'] = (DrugDeathRatioFinal['CountofDrugDeaths']/DrugDeathRatioFinal['CountofDeaths']*100.00)

#Build out Dr Prescribing Ratio Table and Calculate Pescribing Doctor Ratio
PrescriberRateByStateAndYear = carecaid_clean.filter(['Year_State','State','Total Prescribers','Total Opioid Prescribers'])
PrescriberRateByStateAndYear['DrPrescribingOpioidsRatio'] = (PrescriberRateByStateAndYear['Total Opioid Prescribers']/PrescriberRateByStateAndYear['Total Prescribers']*100.00)

PrescriberRateByStateAndYear

#Final Merge two DataFrames
FinalData = pd.merge(DrugDeathRatioFinal, PrescriberRateByStateAndYear, how = "outer", on = ['Year_State'])
FinalData = FinalData.loc[(FinalData["State"] != "National")]

#Setting Plot Values
PrescribingDrRatio = FinalData['DrPrescribingOpioidsRatio']
CountofDrugDeaths = FinalData['DrugDeathRatio']

CountofDrugDeaths
```

```python
#ScatterPlot Datapoints
plt.scatter(PrescribingDrRatio, CountofDrugDeaths)
plt.xlabel("Prescribing Dr Ratio")
plt.ylabel("Drug Death Ratio")
plt.title("Prescribing Doctor Ratio By State versus Drug Death Ratio")
plt.show()
```

# Medicare vs. Medicaid - Highest/Lowest/National Drug Prescription Rates

```python
# create list of years
years = [2015, 2016, 2017, 2018, 2019]

# create list for medicare and medicaid prescription rate max and min
care_prvmax = []
caid_prvmax = []
care_prvmin = []
caid_prvmin = []

# shorten opioid prescribing rate column call 
carerate = carecaid_clean["Opioid Prescribing Rate: Medicare"]
caidrate = carecaid_clean["Opioid Prescribing Rate: Medicaid"]
```

```python
# create df for each year
for year in years:
    yeardf = carecaid_clean.loc[carecaid_clean["Year"]==year]
    
    # add the HIGHEST opioid prescribing rate to list
    care_prvmax.append(yeardf["Opioid Prescribing Rate: Medicare"].max())
    caid_prvmax.append(yeardf["Opioid Prescribing Rate: Medicaid"].max())
    
    # add the LOWEST opioid prescribing rate to list
    care_prvmin.append(yeardf["Opioid Prescribing Rate: Medicare"].min())
    caid_prvmin.append(yeardf["Opioid Prescribing Rate: Medicaid"].min())
```

## Highest Prescribing Rate Data: 

```python
# loop through lists and find the full row for each, per year
# for medicare prescribing rate MAX
for rate in care_prvmax:
    care_top15 = carecaid_clean.loc[carerate == care_prvmax[0]]
    care_top16 = carecaid_clean.loc[carerate == care_prvmax[1]]
    care_top17 = carecaid_clean.loc[carerate == care_prvmax[2]]
    care_top18 = carecaid_clean.loc[carerate == care_prvmax[3]]
    care_top19 = carecaid_clean.loc[carerate == care_prvmax[4]]

# for medicaid prescribing rate MAX
for rate in caid_prvmax:
    caid_top15 = carecaid_clean.loc[caidrate == caid_prvmax[0]]
    caid_top16 = carecaid_clean.loc[caidrate == caid_prvmax[1]]
    caid_top17 = carecaid_clean.loc[caidrate == caid_prvmax[2]]
    caid_top18 = carecaid_clean.loc[caidrate == caid_prvmax[3]]
    caid_top19 = carecaid_clean.loc[caidrate == caid_prvmax[4]]
    
# create individual dfs for the top prescription rates & reset index
care_toppr = pd.concat([care_top15, care_top16, care_top17, care_top18, care_top19]).reset_index().drop(columns=["index"])
caid_toppr = pd.concat([caid_top15, caid_top16, caid_top17, caid_top18, caid_top19]).reset_index().drop(columns=["index"])

# set new index to year and state values
care_toppr["Year: State"] = care_toppr["Year"].map(str) + ": " + care_toppr["State"]
caid_toppr["Year: State"] = caid_toppr["Year"].map(str) + ": " + caid_toppr["State"]

# create condensed individual dfs for medicare and medicaid
care_toppr2 = care_toppr[["Year: State", "Opioid Prescribing Rate: Medicare"]]
caid_toppr2 = caid_toppr[["Year: State", "Opioid Prescribing Rate: Medicaid"]]

# show medicare df
caid_toppr2
```

## Lowest Prescribing Rate Data: 

```python
# loop through lists and find the full row for each, per year
# for medicare prescribing rate MIN
for rate in care_prvmin:
    care_bot15 = carecaid_clean.loc[carerate == care_prvmin[0]]
    care_bot16 = carecaid_clean.loc[carerate == care_prvmin[1]]
    care_bot17 = carecaid_clean.loc[carerate == care_prvmin[2]]
    care_bot18 = carecaid_clean.loc[carerate == care_prvmin[3]]
    care_bot19 = carecaid_clean.loc[carerate == care_prvmin[4]]
    
# for medicaid prescribing rate MIN
for rate in caid_prvmin:
    caid_bot15 = carecaid_clean.loc[caidrate == caid_prvmin[0]]
    caid_bot16 = carecaid_clean.loc[caidrate == caid_prvmin[1]]
    caid_bot17 = carecaid_clean.loc[caidrate == caid_prvmin[2]]
    caid_bot18 = carecaid_clean.loc[caidrate == caid_prvmin[3]]
    caid_bot19 = carecaid_clean.loc[caidrate == caid_prvmin[4]]    
      
# create individual dfs for the top prescription rates & reset index
care_botpr = pd.concat([care_bot15, care_bot16, care_bot17, care_bot18, care_bot19]).reset_index().drop(columns=["index"])
caid_botpr = pd.concat([caid_bot15, caid_bot16, caid_bot17, caid_bot18, caid_bot19]).reset_index().drop(columns=["index"])

# set new index to year and state values
care_botpr["Year: State"] = care_botpr["Year"].map(str) + ": " + care_botpr["State"]
caid_botpr["Year: State"] = caid_botpr["Year"].map(str) + ": " + caid_botpr["State"]

# create condensed individual dfs for medicare and medicaid
care_botpr2 = care_botpr[["Year: State", "Opioid Prescribing Rate: Medicare"]]
caid_botpr2 = caid_botpr[["Year: State", "Opioid Prescribing Rate: Medicaid"]]

# show medicare df
caid_botpr2
```

## CO vs. VA Rate of Change Graph: 

```python
# create fig and subplot function so multiple graphs can be plotted together
fig, ax = plt.subplots(2, figsize=(15,15))

# create x and y variables 
x = [2015, 2016, 2017, 2018, 2019]
co_y1 = colorado["Opioid Prescribing Rate, 1 Year Change: Medicare"]
va_y1 = virginia["Opioid Prescribing Rate, 1 Year Change: Medicare"]
co_y2 = colorado["Opioid Prescribing Rate, 1 Year Change: Medicaid"]
va_y2 = virginia["Opioid Prescribing Rate, 1 Year Change: Medicaid"]

# create variable for scatter plot point size
s = 100

# Medicare scatter plot
ax[0].scatter(x, y = co_y1, s=s)        # Colorado scatter plot
ax[0].scatter(x, y = va_y1, s=s)        # Virginia scatter plot
ax[0].set_title("Medicare", size = 15)  # graph title
ax[0].set_ylabel("", size = 15)         # graph y-axis label
ax[0].plot(x, co_y1)                       # Colorado connecting line 
ax[0].plot(x, va_y1)                       # Virginia connecting line
ax[0].legend(["Colorado", "Virginia"],  # graph legend
             prop = {"size": 10})

# Medicaid scatter plot
ax[1].scatter(x, y = co_y2, s=s)   # Colorado scatter plot
ax[1].scatter(x, y = va_y2, s=s)   # Virginia scatter plot
ax[1].set_title("Medicaid", size = 15)  # graph title
ax[1].set_ylabel("", size = 15)         # graph y-axis label
ax[1].plot(x, co_y2)                    # Colorado connecting line
ax[1].plot(x, va_y2)                    # Virginia connecting line
ax[1].legend(["Colorado", "Virginia"],  # graph legend
             prop = {"size": 10})

# set figure y-axis label, xticks for both graphs, and add grids
for ax in ax.flat:
    ax.set(ylabel = "Change (%)")
    ax.set(xticks = x)
    ax.grid()          

# add title to figure
fig.suptitle("Rate of Change in Opioid Prescription Rates - Colorado vs. Virginia", size = 25)

# save figure to png
plt.savefig("Graphs/ROC_PR_COvsVA.png");
```
